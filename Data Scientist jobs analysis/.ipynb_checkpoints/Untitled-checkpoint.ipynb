```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
import re
import matplotlib.pyplot as plt

joblink2='https://www.indeed.com/viewjob?jk=d11d9853c2d7ffde'

jobtitle_list=[]
companyname_list=[]
location_list=[]
Publication_req_flag_list=[]
DL_flag_list=[]
PhD_masters_flag_list=[]
other_lang_flag_list=[]
Python_flag_list=[]
Spark_flag_list=[]
Visualization_flag_list=[]
R_flag_list=[]
SAS_flag_list=[]
SQL_flag_list=[]
ML_flag_list=[]
four_years_flag_list=[]
two_years_flag_list=[]
entrylevel_flag_list=[]
five_years_flag_list=[]
seven_years_flag_list=[]
ten_years_flag_list=[]
NLP_flag_list=[]
Scrapping_flag_list=[]
Top_mgmt_flag_list=[]
research_flag_list=[]
Senior_flag_list=[]
entrylevel_flag_list=[]
three_years_flag_list=[]

jobtitle_list=list(jobtitle_list)
companyname_list=list(companyname_list)
location_list=list(location_list)



def job_description():

 
    html_page = requests.get(joblink2).text
#         print (html_page)

    html_text = BeautifulSoup(html_page, 'html.parser')
#         print(html_text)

    job = html_text.findAll("h3", {"class": "icl-u-xs-mb--xs icl-u-xs-mt--none jobsearch-JobInfoHeader-title"})
    jobtitle = re.search('none jobsearch-JobInfoHeader-title">(.*)</h3>]', str(job))
    jobtitle=jobtitle.group(1)

    comp_name= html_text.find_all(attrs={"class": "icl-u-lg-mr--sm icl-u-xs-mr--xs"})
    companyname = re.search('icl-u-lg-mr--sm icl-u-xs-mr--xs">(.*)</div>, <div class="icl-u-lg-mr--sm icl', str(comp_name))
    companyname=companyname.group(1)

    location_info=html_text.head.title
    location_info = re.search('\s-\s(.*)\s-\s', str(location_info))
    location_info=location_info.group(1)



    jobtitle=str(jobtitle)
    companyname=str(companyname)
    location_info=str(location_info)


    discription=html_text.findAll('div',{"class": "jobsearch-JobComponent-description icl-u-xs-mt--md"})
    discription=str(discription)
    soup2 = BeautifulSoup(discription, 'html.parser')
    text=soup2.find_all('li')
    text=str(text)
    text=text.replace('</li>','')
    text=text.replace('<li>','')
    print(text)
    stop_words = set(stopwords.words('english'))
    text2=word_tokenize(str(text))

    text_list = []
    punctuation=['(' ,')' ,"'" ,'.' ,',' ,'<' ,'>' ,'[', ']', '{', '}','/b','/p',':','&',';','p','b','#','i','li','/li','']
    for w in text2:
        if w not in stop_words:
            if w not in punctuation:
                text_list.append(w)
    
    Publication_req_flag=0
    DL_flag=0
    PhD_masters_flag=0
    other_lang_flag=0
    Python_flag=0
    Spark_flag=0
    Visualization_flag=0
    R_flag=0
    SAS_flag=0
    SQL_flag=0
    ML_flag=0
    Spark_flag=0
    four_years_flag=0
    three_years_flag=0
    two_years_flag=0
    entrylevel_flag=0
    five_years_flag=0
    seven_years_flag=0
    ten_years_flag=0
    NLP_flag=0
    Scrapping_flag=0
    Top_mgmt_flag=0
    research_flag=0
    Senior_flag=0
    entrylevel_flag=0
    
    
    print (text_list)
    pattern=['SIGKDD','ICML','NIPS']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                Publication_req_flag=1


    pattern=['TensorFlow','Caffe','Keras']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                DL_flag=1


    pattern=['PhD','phd','Ph.D.','Ph.D','ph.d','Masters','masters',"Masters's","masters's"]
    for pat in pattern:
        for text in text_list:
            if pat==text:
                PhD_masters_flag=1


    pattern=['c++','C++','Java','Scala']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                other_lang_flag=1


    pattern=['Python','python']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                Python_flag=1

    pattern=['Spark','spark']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                Spark_flag=1


    pattern=['Tableau','Chartio','QlikSense','Qlikview','ggplot2','plotly','matplotlib','Matplotlib','seaborn','Seaborn']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                Visualization_flag=1 


    pattern=['R']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                R_flag=1
 

    pattern=['SAS']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                SAS_flag=1

    pattern=['SQL']
    for pat in pattern:
        for text in text_list:
            if pat==text:
                SQL_flag=1

    

    patternsML = ['Machine learning','machine learning']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            ML_flag=1


    patternsML = ['Apache spark','apache spark']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            Spark_flag=1


    patternsML = ['4+ years','4+ Years']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            four_years_flag=1


    patternsML = ['3+ years','3+ Years']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            three_years_flag=1


    patternsML = ['2+ years','2+ Years']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            two_years_flag=1


    patternsML = ['0+ years','0+ Years']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            entrylevel_flag=1



    patternsML = ['5+ years','5+ Years']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            five_years_flag=1


    patternsML = ['7+ years','7+ Years']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            seven_years_flag=1


    patternsML = ['10+ years','10+ Years']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            ten_years_flag=1


    patternsML = ['Natural Language Processing','natural language processing','NLP']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            NLP_flag=1


    patternsML = ['Deep learning','deep learning']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            DL_flag=1


    patternsML = ['web scrapping','Web scrapping']
    for pattern in patternsML:
        if re.search(pattern, text) is not None:
            Scrapping_flag=1

    patternsML = ['principle','Principle','chief','Chief','Director','director','VP','Vice president','vice president','president','President']
    for pattern in patternsML:
        if re.search(pattern, jobtitle) is not None:
            Top_mgmt_flag=1
 


    patternsML = ['Research','research']
    for pattern in patternsML:
        if re.search(pattern, jobtitle) is not None:
            research_flag=1
  

    patternsML = ['Senior','senior']
    for pattern in patternsML:
        if re.search(pattern, jobtitle) is not None:
            Senior_flag=1
        

    patternsML = ['intern','Intern','Internship','internship']
    for pattern in patternsML:
        if re.search(pattern, jobtitle) is not None:
            entrylevel_flag=1
      

    jobtitle=str(jobtitle)
    companyname=str(companyname)
    location_info=str(location_info)

    jobtitle_list.append(jobtitle)
    companyname_list.append(companyname)
    location_list.append(location_info)

    Publication_req_flag_list.append(Publication_req_flag)
    DL_flag_list.append(DL_flag)
    PhD_masters_flag_list.append(PhD_masters_flag)
    other_lang_flag_list.append(other_lang_flag)
    Python_flag_list.append(Python_flag)
    Spark_flag_list.append(Spark_flag)
    Visualization_flag_list.append(Visualization_flag)
    R_flag_list.append(R_flag)
    SAS_flag_list.append(SAS_flag)
    SQL_flag_list.append(SQL_flag)
    ML_flag_list.append(ML_flag)
    three_years_flag_list.append(three_years_flag)
    four_years_flag_list.append(four_years_flag)
    two_years_flag_list.append(two_years_flag)
    five_years_flag_list.append(five_years_flag)
    seven_years_flag_list.append(seven_years_flag)
    ten_years_flag_list.append(ten_years_flag)
    NLP_flag_list.append(NLP_flag)
    Scrapping_flag_list.append(Scrapping_flag)
    Top_mgmt_flag_list.append(Top_mgmt_flag)
    research_flag_list.append(research_flag)
    Senior_flag_list.append(Senior_flag)
    entrylevel_flag_list.append(entrylevel_flag)

    df = pd.DataFrame({ 'Job title': jobtitle_list,
                       'Job link': joblink2,
                       'location':location_list,
                       'Company_name': companyname_list,
                       'Publication_req_flag': Publication_req_flag_list,
                       'DL_flag': DL_flag_list,
                       'PhD_masters_flag': PhD_masters_flag_list,
                       'other_lang_flag': other_lang_flag_list,
                       'Python_flag': Python_flag_list,
                       'Spark_flag': Spark_flag_list,
                       'Visualization_flag': Visualization_flag_list,
                       'R_flag': R_flag_list,
                       'SAS_flag': SAS_flag_list,
                       'SQL_flag': SQL_flag_list,
                       'ML_flag': ML_flag_list,
                       'three_years_flag':three_years_flag_list,
                       'four_years_flag': four_years_flag_list,
                       'two_years_flag': two_years_flag_list,
                       'entrylevel_flag': entrylevel_flag_list,
                       'five_years_flag': five_years_flag_list,
                       'seven_years_flag': seven_years_flag_list,
                       'ten_years_flag': ten_years_flag_list,
                       'NLP_flag': NLP_flag_list,
                       'Scrapping_flag': Scrapping_flag_list,
                       'Top_mgmt_flag': Top_mgmt_flag_list,
                       'research_flag': research_flag_list,
                       'Senior_flag': Senior_flag_list,
                       'entrylevel_flag': entrylevel_flag_list})       
    return df

JD=job_description()
```
